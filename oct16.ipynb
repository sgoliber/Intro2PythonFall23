```python
import pandas as pd 
```

```python
data = pd.read_csv('d18o_datafile.csv')
data = data.rename(columns={"DAY": "days", data.columns[1]: "d18o"})
data
```

```python
fig, ax = plt.subplots()

ax.plot(data['days'],data['d18o'])
```

```python
above_neg14 = data[data["d18o"] > -14]

fig, axs = plt.subplots(2) 
axs[0].plot(data.days,data.d18o)
axs[1].plot(above_neg14.days,above_neg14.d18o)
```

```python
fig, ax = plt.subplots()

ax.plot(data.days,data.d18o)
ax.set_title('Î´18O over time')
ax.set_xlabel('Days')
ax.set_ylabel('d180o')
```

```python

fig, (ax1,ax2) = plt.subplots(2) 

ax1.scatter(data.days,data.d18o)
ax1.set_title('Î´18O over time')
ax1.set_xlabel('Days')
ax1.set_ylabel('d180o')


ax2.plot(above_neg14.days,above_neg14.d18o)
ax2.set_title('Î´18O over time')
ax2.set_xlabel('Days')
ax2.set_ylabel('d180o')
```

```python
#To precisely control the positioning of the subplots, we can use fig.subplots_adjust
# expressed as a fraction of the average axis height

fig, (ax1,ax2) = plt.subplots(2) 
fig.subplots_adjust(hspace=.5)

ax1.plot(data.days,data.d18o)
ax1.set_title('Î´18O over time')
ax1.set_xlabel('Days')
ax1.set_ylabel('d180o')


ax2.plot(above_neg14.days,above_neg14.d18o)
ax2.set_title('Î´18O over time')
ax2.set_xlabel('Days')
ax2.set_ylabel('d180o')
```

```python
fig, ax = plt.subplots()

ax.plot(data.days,data.d18o,label='All Data',zorder=0)

ax.scatter(above_neg14.days,above_neg14.d18o, color = 'purple',label='Selected data',zorder=1)

ax.set_title('Î´18O over time')
ax.set_xlabel('Days')
ax.set_ylabel('d180o')

ax.legend()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (10,5)) 
fig.subplots_adjust(wspace=.5)

#ax 1
ax1.plot(data.days,data.d18o)
ax1.scatter(above_neg14.days,above_neg14.d18o, color = 'purple',label='Selected data')
ax1.set_title('Î´18O over time')
ax1.set_xlabel('Days')
ax1.set_ylabel('d180o')

#ax 2
ax2.hist(data.d18o, bins=5)
ax2.set_title('Î´18O Histogram')
ax2.set_xlabel('value')
ax2.set_ylabel('count')

fig.suptitle('Î´18O Test Data') # or plt.suptitle('Main title')
```

```python
import numpy as np
x = np.linspace(0, 2, 100)  

fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(x, x, label='linear')  
ax.plot(x, x**2, label='quadratic')  
ax.plot(x, x**3, label='cubic')  
ax.set_xlabel('x label')  
ax.set_ylabel('y label')  
ax.set_title("Simple Plot")  
ax.legend()
```

```python
x = np.linspace(0, 2, 100)  
plt.figure(figsize=(5, 2.7))
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```
